> 本次实验使用了单因素多组对照实验进行分析。首先我们使用计算公式1对模型进行评估，测试不同populations 与 generations 是否对模型有影响。


> 由于实验条件限制，我们只能在cpu环境下进行。 每次模型训练都需要数个小时，gens作为单因素对照试验中只选用了两组，一组是2000pops 与 100gens，一组为2000pops 与 1000gens。尝试在1000gens中寻找到最优模型。我们可以从图中看出，gens对于模型训练起到xxx的作用。。。。。


>图2是同为迭代100次，不同的population训练的模型结果。从图可知，population1000的模型训练结果最好，pops1500的训练最差。 根据数据结果可知。模型训练好坏与propulsion无关。
>与此同时，我们可以看到ftness的值与迭代次数不存在线性相关。在某一次的迭代中突然fitness升高，这是在不断地基因重组与突变的过程中，随机的生成了较好的参数，snake学习到较优的模型。
>这也印证了遗传算法的不稳定性与随机性。从图中可以看到，fitness达到峰值后也存在较大的波动，即后代基因重组与突变的结果并一定比之前的好。这也符合遗传算法的特性。



> 上述实验都是基于计算公式1进行模型评估。在此我们又试图调整fitness的评价计算公式2进行模型训练。在此仅用同为迭代100次，不同的population进行模型训练。可以从图3看出。。。。


> 由于评价标准的改变，两种评价公式训练出的最优模型无法从数据上判定好坏。然而模型测试的环节之中，苹果出现的位置是随机的，这导致模型每次测试都存在不确定性。两种评价机制的优劣只能通过人为主观的模型测试进行判定。所以在不同的模型评价机制下，无法客观评判遗传算法模型的优劣。
