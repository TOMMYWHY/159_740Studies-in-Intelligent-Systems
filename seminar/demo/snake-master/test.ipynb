# Result analysis

````python
import matplotlib
from matplotlib import pyplot as plt
from matplotlib.pyplot import GridSpec
import numpy as np
import pandas as pd
import os, sys
from pylab import rcParams
# rcParams['figure.figsize'] = 25, 20

rcParams['font.size'] = 20
rcParams['axes.facecolor'] = 'white'
matplotlib.style.use('ggplot')
%matplotlib inline
````

````python
col_names= ["pop_size",            "gen_num",             "fitness",             "avg_top6",                  "avg_last6"]
````

````python
df_500 = pd.read_csv("./best_model/pops500_gen100/output_pops500_gn100.csv",names=col_names)
df_1000 = pd.read_csv("./best_model/pops1000_gen100/output_pops1000_gn100.csv",names=col_names)
df_1500 = pd.read_csv("./best_model/pops1500_gen100/output_pops1500_gn100.csv",names=col_names) 
df_2000 = pd.read_csv("./best_model/pops2000_gen100/output_pops2000_gn100.csv",names=col_names) 

````

````python
#max fitness
print(df_500.iloc[df_500['fitness'].argmax()].tolist())
print(df_1000.iloc[df_500['fitness'].argmax()].tolist())
print(df_1500.iloc[df_500['fitness'].argmax()].tolist())
print(df_2000.iloc[df_500['fitness'].argmax()].tolist())
````

````python
# df_500[['fitness']]
plt.figure(figsize=(6,4))
plt.figure(dpi=150)
plt.plot(df_500["gen_num"],df_500["fitness"], c="red",label="pops 500")
plt.plot(df_1000["gen_num"],df_1000["fitness"], c="hotpink",label="pops 1000")
plt.plot(df_1500["gen_num"],df_1500["fitness"], c="darkseagreen",label="pops 1500")
plt.plot(df_2000["gen_num"],df_2000["fitness"], c="blueviolet",label="pops 2000")


plt.ylabel(u"GA performance ") 
plt.tight_layout()
# plt.title("Fitness ", alpha=0.8)
plt.legend(loc='upper left')
plt.savefig("Fitness.png")
plt.show()
````

# User guide for the demonstration codes


1. Dependencies
Using Python 3.6 or uper environment.
The project needs libraries: joblib,numpy,numba,pygame
2. User guide 
2.1 demonstration 
execute 'run.py' with parameter "--mode demo" will loading the 1000 population and 100generation model. 

``` shell
 python run.py --mode demo

```
If you prefer loading different model have to modify the path in main.py file.
-img

If training snakes please select parameter "--mode train" and need to setup population and generations as below.

``` shell
 python run.py --mode train --pops 1000 --gens 100

```
Using different fitness format have to edit in the file of snake.py.
-img

we still prvide mode of "play" for user slef play as well. 
``` shell
 python run.py --mode play
```
